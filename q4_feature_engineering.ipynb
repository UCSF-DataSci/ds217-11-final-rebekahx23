## Objective

Create derived features, perform time-based aggregations, and calculate rolling windows for time series analysis.

**Time Series Note:** Rolling windows are essential for time series data. They capture temporal dependencies (e.g., 7-hour rolling mean captures short-term patterns). Use pandas `rolling()` method with `window` parameter to specify the number of periods.

## Your Approach

1. **Create derived features** - Differences, ratios, interactions between variables (watch for division by zero)
2. **Calculate rolling windows** - Use `.rolling()` on predictor variables to capture temporal patterns

   ⚠️ **Data Leakage Warning:** Do not create ANY features that use your target variable - this includes rolling windows, differences, ratios, or interactions involving the target. For example, if predicting Air Temperature, do not create `air_temp * humidity` or `air_temp - wet_bulb`. Only derive features from other predictor variables.

3. **Create categorical features** - Bin continuous variables if useful (optional)
4. **Check for infinity values** - Ratios can produce infinity; replace with NaN and handle appropriately
5. **Document and save** - Remember to `reset_index()` before saving CSVs

---

## Decision Points

- **Derived features:** What relationships might be useful? Temperature differences? Ratios? Interactions between variables?
- **Rolling windows:** What window size makes sense? 7 hours? 24 hours? Consider the temporal scale of your data. For hourly data, 7-24 hours captures daily patterns.
- **Time-based aggregations:** Aggregate by hour? Day? Week? What temporal granularity is useful for your analysis?

````python
# part 1

df = pd.read_csv('output/q3_wrangled_data.csv')
print(df.columns)
df['Measurement Timestamp'] = pd.to_datetime(df['Measurement Timestamp'])
dt = df["Measurement Timestamp"]
df["Hour"] = dt.dt.hour
df["DayOfWeek"] = dt.dt.dayofweek          # 0 = Monday
df["Month"] = dt.dt.month
df["IsWeekend"] = (dt.dt.dayofweek >= 5).astype(int)
df["IsNight"] = ((dt.dt.hour < 6) | (dt.dt.hour >= 18)).astype(int)

df["AirTemp_Change"] = df["Air Temperature"].diff()
df["Humidity_Change"] = df["Humidity"].diff()
df["Pressure_Change"] = df["Barometric Pressure"].diff()
df["WindSpeed_Change"] = df["Wind Speed"].diff()

df["Wind_Gust_Ratio"] = df["Maximum Wind Speed"] / df["Wind Speed"].replace(0, np.nan)

# Rain per interval (if Interval Rain is > 0)
df["Rain_Per_Interval"] = df["Total Rain"] / df["Interval Rain"].replace(0, np.nan)

# “humidity-normalized” temperature
df["Temp_to_Humidity_Ratio"] = df["Air Temperature"] / df["Humidity"].replace(0, np.nan)

# Temp × humidity → crude “heat load” interaction
df["Temp_Humidity_Interaction"] = df["Air Temperature"] * df["Humidity"]

# Wind × solar → windy & sunny conditions interaction
df["Wind_Solar_Interaction"] = df["Wind Speed"] * df["Solar Radiation"]

# Binary flag: any rain happening?
df["IsRaining"] = (df["Rain Intensity"] > 0).astype(int)

# Save to CSV without index
df.to_csv('output/q4_features.csv', index=False)
print("Saved derived features to output/q4_features.csv")

````

````python
### Part 2

df["Measurement Timestamp"] = pd.to_datetime(df["Measurement Timestamp"], errors="coerce")
df = df.set_index("Measurement Timestamp")
df = df.sort_values("Measurement Timestamp")
df["wind_speed_rolling_7h"] = df["Wind Speed"].rolling("7h", min_periods=1).mean()
df["humidity_rolling_24h"] = df["Humidity"].rolling("24h", min_periods=1).mean()
df["pressure_rolling_7h"] = df["Barometric Pressure"].rolling("7h", min_periods=1).mean()
df.reset_index().to_csv('output/q4_rolling_features.csv', index=False)
print("Saved rolling window features to output/q4_rolling_features.csv")    
````

````python
### Part 3

with open("output/q4_feature_list.txt", "w") as f:
    f.write("Hour\n")
    f.write("DayOfWeek\n")
    f.write("Month\n")
    f.write("IsWeekend\n")
    f.write("IsNight\n")
    f.write("AirTemp_Change\n")
    f.write("Humidity_Change\n")
    f.write("Pressure_Change\n")
    f.write("WindSpeed_Change\n")
    f.write("Wind_Gust_Ratio\n")
    f.write("Rain_Per_Interval\n")
    f.write("Temp_to_Humidity_Ratio\n")
    f.write("Temp_Humidity_Interaction\n")
    f.write("Wind_Solar_Interaction\n")
    f.write("IsRaining\n")
    f.write("wind_speed_rolling_7h\n")
    f.write("humidity_rolling_24h\n")
    f.write("pressure_rolling_7h\n")
print("Saved feature list to output/q4_feature_list.txt")
````
