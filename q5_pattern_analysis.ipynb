````python
### Part 1
key_vars = [
    "Air Temperature", 
    "Humidity", 
    "Wind Speed", 
    "Barometric Pressure",
    "Solar Radiation"
]

corr_matrix = df[key_vars].corr()

corr_matrix.to_csv("output/q5_correlations.csv")

print("Saved subset correlation matrix to output/q5_correlations.csv")

````

````python
# Part 2 Turn variables into patterns

daily_trend = df[["Air Temperature", "Humidity"]].resample("D").mean()
df["Hour"] = df.index.hour
hourly_pattern = df.groupby("Hour")["Air Temperature"].mean()
sns.set_style("whitegrid")

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle("Q5: Patterns, Trends, and Correlations", fontsize=16)

# Subplot 1: Trend over time (daily means)
ax1 = axes[0, 0]
daily_trend.plot(ax=ax1)
ax1.set_title("Daily Trend: Air Temperature & Humidity")
ax1.set_xlabel("Date")
ax1.set_ylabel("Value")
ax1.legend(title="Variable")

# Subplot 2: Seasonal pattern by hour of day
ax2 = axes[0, 1]
ax2.plot(hourly_pattern.index, hourly_pattern.values, marker="o")
ax2.set_title("Seasonal Pattern: Air Temperature by Hour of Day")
ax2.set_xlabel("Hour of Day")
ax2.set_ylabel("Mean Air Temperature")
ax2.set_xticks(range(0, 24, 2))

# Subplot 3: Correlation heatmap
ax3 = axes[1, 0]
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", ax=ax3)
ax3.set_title("Correlation Heatmap (Key Variables)")

# Turn off unused subplot
axes[1, 1].axis("off")

plt.tight_layout(rect=[0, 0, 1, 0.95])

# Save
plt.savefig("output/q5_patterns.png", dpi=150)
plt.close()

print("Saved advanced pattern visualizations to output/q5_patterns.png")
````
